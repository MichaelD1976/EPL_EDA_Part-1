```python
# remove players who have played less than 45 mins
df = df[df.Mins >=45]
```

```python
# Replace Null values with zeros
df = df.fillna(0)
df.head(10)
```

```python
# add new column
df['CrdY/90'] = round((df['CrdY'] / df['90s']),2)
```

#### Standardize player name characters to latin

```python
# standardize foreign lettering to latin
Name = []
for p_name in df.loc[:,'Player']:
    p_name = unidecode.unidecode(p_name)
    Name.append(p_name)
    
df['Name'] = Name
df = df.drop(['Player'],axis=1)    # drop original column

```

```python
# reorder
df = df[['Team', 'Name', 'Nat', 'Pos', 'Age', 'MP', 'Starts', 'Mins', '90s',
       'Gls', 'Gls/90', 'Ast', 'Ast/90', 'PK', 'CrdY', 'CrdY/90', 'CrdR', 'xG', 'xG/90', 'Sh',
       'SoT', 'SoT%', 'Sh/90', 'SoT/90', 'G/Sh', 'G/SoT', 'Dist', 'FK',
      ]]
```

#### Standardize 3 letter Nation ID

```python
# Standardize country to 3 letter code
new_Nat = []
for string in df['Nat']:
    new_string = string[-3:]
    new_Nat.append(new_string)
    
df['Nation'] = new_Nat
df.drop(['Nat'],axis=1, inplace = True)

#df = df['Team'] + df[ ['Name'] + [ col for col in df.column if col != 'Name' and col!= 'Nation' ] ] 
```

```python
# reorder
df = df[['Team', 'Name', 'Nation', 'Pos', 'Age', 'MP', 'Starts', 'Mins', '90s', 'Gls', 'Gls/90', 'xG', 'xG/90', 'Sh', 'Sh/90', 'G/Sh', 'SoT', 'SoT%', 'SoT/90', 'G/SoT', 'Dist', 'FK', 'Ast', 'Ast/90', 'PK', 'CrdY', 'CrdY/90', 'CrdR']]
df
```
